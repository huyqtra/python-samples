```python
# Define constructs and their items
challenge_stressors = ['Challenge_stressor_1', 'Challenge_stressor_2', 'Challenge_stressor_3',
                       'Challenge_stressor_4', 'Challenge_stressor_5']
hindrance_stressors = ['Hindrance_1', 'Hindrance_2', 'Hindrance_3', 'Hindrance_4', 'Hindrance_5']
self_efficacy = ['Self_Efficacy_1R', 'Self_Efficacy_2R', 'Self_Efficacy_3R',
                 'Self_Efficacy_5R', 'Self_Efficacy_4R', 'Self_Efficacy_6R']
unethical_behavior = ['Uneth_1', 'Uneth_2', 'Uneth_3', 'Uneth_4', 'Uneth_5', 'Uneth_6',
                      'Uneth_7', 'Uneth_8', 'Uneth_9']
moral_identity = ['MID_1', 'MID_2', 'MID_3']

# Perform Exploratory Factor Analysis (EFA) for each construct
def efa_analysis(df, items):
    """
    Perform Exploratory Factor Analysis and filter low-loading items.
    """
    fa = FactorAnalyzer(n_factors=1, rotation=None)
    fa.fit(df[items])
    loadings = pd.DataFrame(fa.loadings_, index=items, columns=['Loading'])

    # Filter items with |loading| < 0.5
    filtered_items = loadings[loadings['Loading'].abs() >= 0.5].index.tolist()
    return loadings, filtered_items

# Calculate AVE and CR for each construct
def calculate_ave_cr(loadings):
    """
    Calculate Average Variance Extracted (AVE) and Composite Reliability (CR).
    """
    # Squared loadings
    squared_loadings = loadings['Loading'] ** 2

    # Average Variance Extracted (AVE)
    ave = squared_loadings.mean()

    # Composite Reliability (CR)
    total_loadings = loadings['Loading'].sum()
    total_error_variances = (1 - squared_loadings).sum()
    cr = (total_loadings ** 2) / ((total_loadings ** 2) + total_error_variances)

    return ave, cr

# Calculate Cronbach's Alpha
def calculate_cronbach_alpha(df, items):
    """
    Calculate Cronbach's Alpha for a given set of items.
    """
    df_subset = df[items]
    item_variances = df_subset.var(axis=0, ddof=1)
    total_variance = df_subset.sum(axis=1).var(ddof=1)
    num_items = df_subset.shape[1]
    alpha = num_items / (num_items - 1) * (1 - item_variances.sum() / total_variance)
    return alpha

# Store loadings and filtered items in dictionaries
loadings_dict = {}
filtered_items_dict = {}

# Perform EFA and store results for each construct
constructs = {
    'Challenge Stressors': challenge_stressors,
    'Hindrance Stressors': hindrance_stressors,
    'Self-Efficacy': self_efficacy,
    'Unethical Behavior': unethical_behavior,
    'Moral Identity': moral_identity
}

for name, items in constructs.items():
    loadings, filtered_items = efa_analysis(df, items)
    loadings_dict[name] = loadings
    filtered_items_dict[name] = filtered_items
    print(f"\n{name} Loadings:\n{loadings}")

# Loop through constructs and calculate AVE, CR, and Cronbach's Alpha
for name, filtered_items in filtered_items_dict.items():
    if filtered_items:  # Only calculate if there are items left after filtering
        ave, cr = calculate_ave_cr(loadings_dict[name].loc[filtered_items])
        alpha = calculate_cronbach_alpha(df, filtered_items)
        print(f"\n{name} - AVE: {ave:.2f}, CR: {cr:.2f}, Cronbach's Alpha: {alpha:.2f}")
    else:
        print(f"\n{name} - All items were removed due to low absolute loadings.")

```

```python
df=cleaned_data
# Define constructs and their items
challenge_stressors = ['ChStr_1',
 'ChStr_2',
 'ChStr_3',
 'ChStr_4',
 'ChStr_5']
hindrance_stressors = ['HiStr_1',
 'HiStr_2',
 'HiStr_3',
 'HiStr_4',
 'HiStr_5']
self_efficacy = ['SE_1R',
 'SE_2R',
 'SE_3R',
 'SE_4',
 'SE_5R',
 'SE_6']

moral_identity = ['MD_1',
 'MD_2',
 'MD_3']

unethical_behavior = ['Une_1',
 'Une_2',
 'Une_3',
 'Une_4',
 'Une_5',
 'Une_6',
 'Une_7',
 'Une_8',
 'Une_9']

Self_LedD =['AuDe_1',
 'AuDe_2',
 'AuDe_3',
 'AuDe_4',
 'AuDe_5',]

Peer_LedD =['CoWDev_1',
 'CoWDev_2',
 'CoWDev_3',
 'CoWDev_4']

# Perform Exploratory Factor Analysis (EFA) for each construct
def efa_analysis(df, items):
    """
    Perform Exploratory Factor Analysis and filter low-loading items.
    """
    fa = FactorAnalyzer(n_factors=1, rotation=None)
    fa.fit(df[items])
    loadings = pd.DataFrame(fa.loadings_, index=items, columns=['Loading'])

    # Filter items with |loading| < 0.5
    filtered_items = loadings[loadings['Loading'].abs() >= 0.5].index.tolist()
    return loadings, filtered_items

# Calculate AVE and CR for each construct
def calculate_ave_cr(loadings):
    """
    Calculate Average Variance Extracted (AVE) and Composite Reliability (CR).
    """
    # Squared loadings
    squared_loadings = loadings['Loading'] ** 2

    # Average Variance Extracted (AVE)
    ave = squared_loadings.mean()

    # Composite Reliability (CR)
    total_loadings = loadings['Loading'].sum()
    total_error_variances = (1 - squared_loadings).sum()
    cr = (total_loadings ** 2) / ((total_loadings ** 2) + total_error_variances)

    return ave, cr

# Calculate Cronbach's Alpha
def calculate_cronbach_alpha(df, items):
    """
    Calculate Cronbach's Alpha for a given set of items.
    """
    df_subset = df[items]
    item_variances = df_subset.var(axis=0, ddof=1)
    total_variance = df_subset.sum(axis=1).var(ddof=1)
    num_items = df_subset.shape[1]
    alpha = num_items / (num_items - 1) * (1 - item_variances.sum() / total_variance)
    return alpha

# Store loadings and filtered items in dictionaries
loadings_dict = {}
filtered_items_dict = {}

# Perform EFA and store results for each construct
constructs = {
    'Challenge Stressors': challenge_stressors,
    'Hindrance Stressors': hindrance_stressors,
    'Self-Efficacy': self_efficacy,
    'Moral Identity': moral_identity,
    'Unethical Behavior': unethical_behavior,
    'SelfLed Development': Self_LedD,
    'PeerLed Development': Peer_LedD
}

for name, items in constructs.items():
    loadings, filtered_items = efa_analysis(df, items)
    loadings_dict[name] = loadings
    filtered_items_dict[name] = filtered_items
    print(f"\n{name} Loadings:\n{loadings}")

# Loop through constructs and calculate AVE, CR, and Cronbach's Alpha
for name, filtered_items in filtered_items_dict.items():
    if filtered_items:  # Only calculate if there are items left after filtering
        ave, cr = calculate_ave_cr(loadings_dict[name].loc[filtered_items])
        alpha = calculate_cronbach_alpha(df, filtered_items)
        print(f"\n{name} - AVE: {ave:.2f}, CR: {cr:.2f}, Cronbach's Alpha: {alpha:.2f}")
    else:
        print(f"\n{name} - All items were removed due to low absolute loadings.")
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler
from pingouin import mediation_analysis

# Step 1: Standardize Variables
df_standardized = df.copy()
scaler = StandardScaler()
standardized_columns = ['ChStr', 'HiStr', 'AuDe', 'CowDev', 'MoId', 'Age', 'Gender', 'Ethnic', 'SeEff', 'Uneth']
df_standardized[standardized_columns] = scaler.fit_transform(df_standardized[standardized_columns])

# Step 2: Create Interaction Terms for Standardized Variables
interaction_terms = {
    'ChStr_x_AuDe': ('ChStr', 'AuDe'),
    'HiStr_x_AuDe': ('HiStr', 'AuDe'),
    'ChStr_x_CowDev': ('ChStr', 'CowDev'),
    'HiStr_x_CowDev': ('HiStr', 'CowDev'),
}

for term, (var1, var2) in interaction_terms.items():
    df_standardized[term] = df_standardized[var1] * df_standardized[var2]

# Step 3: Moderated Mediation Analysis with Standardized Coefficients
def moderated_mediation_analysis_standardized(X_var, M_var, Y_var, moderator_var, interaction_term, controls, control_var):
    """
    Perform moderated mediation analysis, calculate standardized coefficients,
    and report moderated mediation index scores with bootstrapped CIs.
    """
    # Step 3a: Mediator Model (X -> M)
    mediator_model = sm.OLS(
        df_standardized[M_var],
        sm.add_constant(df_standardized[[X_var, moderator_var, interaction_term, control_var] + controls])
    ).fit()

    # Step 3b: Outcome Model (X, M -> Y)
    outcome_model = sm.OLS(
        df_standardized[Y_var],
        sm.add_constant(df_standardized[[X_var, M_var, moderator_var, interaction_term, control_var] + controls])
    ).fit()

    # Step 3c: Mediation Analysis for Indirect Effects
    mediation_results = mediation_analysis(
        data=df_standardized,
        x=X_var,
        m=M_var,
        y=Y_var,
        covar=controls + [control_var],
        alpha=0.05,
        n_boot=5000
    )

    # Step 3d: Calculate Moderated Mediation Index
    interaction_coefficient = mediator_model.params[interaction_term]
    mediator_to_outcome_coefficient = outcome_model.params[M_var]
    moderated_mediation_index = interaction_coefficient * mediator_to_outcome_coefficient

    # Bootstrapping for Confidence Intervals
    bootstrapped_indices = []
    for _ in range(5000):
        sample = df_standardized.sample(frac=1, replace=True)
        boot_mediator_model = sm.OLS(
            sample[M_var],
            sm.add_constant(sample[[X_var, moderator_var, interaction_term, control_var] + controls])
        ).fit()
        boot_outcome_model = sm.OLS(
            sample[Y_var],
            sm.add_constant(sample[[X_var, M_var, moderator_var, interaction_term, control_var] + controls])
        ).fit()
        boot_interaction_coefficient = boot_mediator_model.params[interaction_term]
        boot_mediator_to_outcome_coefficient = boot_outcome_model.params[M_var]
        boot_index = boot_interaction_coefficient * boot_mediator_to_outcome_coefficient
        bootstrapped_indices.append(boot_index)

    ci_lower = np.percentile(bootstrapped_indices, 2.5)
    ci_upper = np.percentile(bootstrapped_indices, 97.5)

    # Report Results
    print(f"\n### Moderated Mediation Analysis ({X_var} -> {M_var} -> {Y_var}) ###")
    print(f"Interaction Term: {interaction_term}")
    print("\nMediator Model Summary:")
    print(mediator_model.summary())
    print("\nOutcome Model Summary:")
    print(outcome_model.summary())
    print("\nMediation Results (Standardized):")
    print(mediation_results)
    print(f"\nModerated Mediation Index: {moderated_mediation_index:.3f}")
    print(f"95% CI: [{ci_lower:.3f}, {ci_upper:.3f}]")

# Step 4: Perform Analyses for All Models
controls = ['MoId', 'Age', 'Gender', 'Ethnic']

# Model 1a: Challenge Stressors (ChStr) as independent variable with Hindrance Stressors (HiStr) as control
moderated_mediation_analysis_standardized(
    X_var='ChStr',
    M_var='SeEff',
    Y_var='Uneth',
    moderator_var='AuDe',
    interaction_term='ChStr_x_AuDe',
    controls=controls,
    control_var='HiStr'
)

# Model 1b: Hindrance Stressors (HiStr) as independent variable with Challenge Stressors (ChStr) as control
moderated_mediation_analysis_standardized(
    X_var='HiStr',
    M_var='SeEff',
    Y_var='Uneth',
    moderator_var='AuDe',
    interaction_term='HiStr_x_AuDe',
    controls=controls,
    control_var='ChStr'
)

# Model 2a: Challenge Stressors (ChStr) as independent variable with Hindrance Stressors (HiStr) as control
moderated_mediation_analysis_standardized(
    X_var='ChStr',
    M_var='SeEff',
    Y_var='Uneth',
    moderator_var='CowDev',
    interaction_term='ChStr_x_CowDev',
    controls=controls,
    control_var='HiStr'
)

# Model 2b: Hindrance Stressors (HiStr) as independent variable with Challenge Stressors (ChStr) as control
moderated_mediation_analysis_standardized(
    X_var='HiStr',
    M_var='SeEff',
    Y_var='Uneth',
    moderator_var='CowDev',
    interaction_term='HiStr_x_CowDev',
    controls=controls,
    control_var='ChStr'
)
```
